```python
# Q1.Can you find that there has been an improvement in mpg over the years ?
df_1 = data.drop(['cylinders' , 'cubicinches' , 'hp' , 'weightlbs' , 'time-to-60','brand'], axis=1)
df_1main = df_1.groupby(data['year']).mean('mpg')
df_1main
```

```python
#using matplotlib to plot the graph
import matplotlib.pyplot as plt
# plt.bar(df_1main['year'],df_1main['mpg'])
# plt.title('Affect of mpg over years')
# plt.xlabel('Years')
# plt.ylabel('MPG')
# plt.show()

# Using seaborn
import seaborn as sns
# plt.figure(figsize=(4,4))
sns.barplot(x=df_1main['year'],y=df_1main['mpg'],data=df_1main)
```

A1) There is has been a constant increase in the mpg over years.
[Except for the year 1974 (where the value of mpg dropped) and the year 1981 (where the value of mpg hiked)]

```python
#Q2.Does cubic inches of a vehicle affect the vehicle's time to reach 60 Miles per hour?
# df_2 = data.groupby(data['cubicinches']).sum('time-to-60')
# df_1 = data.sort_values(by = 'time-to-60')

df_2 = data.groupby('time-to-60')['cubicinches'].mean()
df_2sort=df_2.to_frame()
df_2sort.reset_index(drop=False,inplace=True)
df_2sort
#Using Scatterplot to plot this.
```

```python
#using matplotlib to plot the graph
import matplotlib.pyplot as plt
# x=df_1main['year']
# y=df_1main['mpg']

plt.bar(df_2sort['time-to-60'],df_2sort['cubicinches'])
plt.title("Does cubicinches affect the vehicle's time to reach 60 Miles per hour? ")
plt.xlabel('time-to-60')
plt.ylabel('cubicinches')
plt.show()
```

A2) Cubicinches and vehicle's time to reach 60 Miles per hour are inversely proportional to each other

```python
#4.How does horsepower affect the average of the vehicle ?
df_4 = data.groupby(data['hp']).sum('mpg')
df_4 = data.drop(['cylinders' , 'brand' ,'cubicinches' , 'weightlbs' , 'year','time-to-60'], axis=1)
df_4sort = df_4.sort_values(by = 'hp')
df_4sort
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_4sort['mpg'],df_4sort['hp'])
plt.title(" Affect of horsepower on the average of the vehicle ")
plt.xlabel('vehicle avg')
plt.ylabel('hp')
plt.show()
```

A4) Horsepower is inversely proportional to mpg. 

    -->If the HP is high, You will get less mileage.
    
    -->If the HP is low, You will get a higher mileage.

Q7.What relation between cubic inches , number of cylinders and horsepower can you observe from the data ?


For this question, We will plot two line graphs between (cubicinches and number of cylinders) & (number of cylinders and horsepower). In the end we'll mutually relate cubicinches to horsepower.



```python
#For cubicinches and number of cylinders

df_7_a = data.groupby('cylinders')['cubicinches'].mean()
df_7_A = df_7_a.to_frame()
df_7_A.reset_index(drop=False,inplace=True)
df_7_A
```

```python
#For horsepower and number of cylinders

df_7_b= data.groupby('cylinders')['hp'].mean()
df_7_B = df_7_b.to_frame()
df_7_B.reset_index(drop=False,inplace=True)
df_7_B
```

```python
sns.lineplot(data=df_7_A, x="cylinders", y="cubicinches")
sns.lineplot(data=df_7_B, x="cylinders", y="hp")
```
